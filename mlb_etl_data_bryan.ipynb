```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_payroll = "Resources/mlb_payroll.csv"
payroll_data_df= pd.read_csv(csv_payroll)
payroll_data_df.head()
```

```python
#cleaning up payroll data
new_payroll_df=payroll_data_df[["Year","Total Payroll","Player Count","Team Count","Highest Payroll","Lowest Payroll"]].copy()
new_payroll_df.head()
```

```python
#importing team data
teams_main_info= "Resources/Teams.csv"
teams_df=pd.read_csv(teams_main_info)
teams_df.head()
```

```python
#team data clean up
teams_info_clean=teams_df[["yearID","lgID","teamID","divID","W","L","name"]].copy()
teams_info_clean
```

```python
#team data from 2000 to 2015 last year of info
teams_2000= teams_info_clean[teams_info_clean.yearID >= 2000]
teams_2000
```

```python
import_main= "Resources/Master.csv"
main_df=pd.read_csv(import_main)
main_df.head()
```

```python
clean_main_df= main_df[["playerID","nameFirst","nameLast"]].copy()
clean_main_df.head(3)
```

```python
import_salaries= "Resources/Salaries.csv"
salaries_df=pd.read_csv(import_salaries)
salaries_df.head(3)
```

```python
main_salaries=pd.merge(clean_main_df,salaries_df,how="inner", on= "playerID")
main_salaries.head()
```

```python
main_sal_2000 =main_salaries[main_salaries.yearID >= 2000].copy()
main_sal_2000.head(3)
```

```python
main_w_l_record_unclean=pd.merge(main_sal_2000,teams_2000, how="inner", on= "teamID")
main_w_l_record_unclean.head(3)
```

```python
main_w_l_record_c=main_w_l_record_unclean.drop(['yearID_x','lgID_x'], axis=1)
```

```python
main_unclean=main_w_l_record_c.rename(columns={'yearID_y':'yearID',
                                               'lgID_y':'lgID',})
```

```python
main_organized=main_unclean[['playerID','nameFirst','nameLast','salary','yearID','teamID','lgID','divID','W','L','name']]
main_organized.head()
```
